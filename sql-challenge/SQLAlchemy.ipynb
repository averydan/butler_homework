```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:<ENTER-PASSWORD>@localhost:5432/employees')
conn = engine.connect()
```

```python
salaries=pd.read_sql('''
    SELECT employees.emp_no, employees.last_name, employees.first_name, employees.sex, salaries.salary
    FROM employees
    LEFT JOIN salaries
    ON salaries.emp_no = employees.emp_no
    ''', conn)
salaries
```

```python
hired_1986=pd.read_sql("SELECT first_name, last_name, hire_date FROM employees WHERE EXTRACT(YEAR FROM employees.hire_date) = 1986", conn)
hired_1986
```

```python
departments=pd.read_sql('''
    SELECT departments.dept_no, departments.dept_name, dept_manager.emp_no, employees.last_name, employees.first_name
    FROM departments
    LEFT JOIN dept_manager
    ON dept_manager.dept_no = departments.dept_no
    LEFT JOIN employees
    ON dept_manager.emp_no = employees.emp_no
    ''', conn)
departments
```

```python
department_employees=pd.read_sql(f'''
SELECT employees.emp_no, employees.last_name, employees.first_name, departments.dept_name
FROM dept_emp
LEFT JOIN employees
ON employees.emp_no = dept_emp.emp_no
LEFT JOIN departments
ON departments.dept_no = dept_emp.dept_no
    ''', conn)
department_employees
```

```python
sales_employees=pd.read_sql(f'''
SELECT employees.emp_no, employees.last_name, employees.first_name, departments.dept_name
FROM dept_emp
LEFT JOIN employees
ON employees.emp_no = dept_emp.emp_no
LEFT JOIN departments
ON departments.dept_no = dept_emp.dept_no
WHERE departments.dept_name = 'Sales'
    ''', conn)
sales_employees
```

```python
sales_dev_employees=pd.read_sql(f'''
SELECT employees.emp_no, employees.last_name, employees.first_name, departments.dept_name
FROM dept_emp
LEFT JOIN employees
ON employees.emp_no = dept_emp.emp_no
LEFT JOIN departments
ON departments.dept_no = dept_emp.dept_no
WHERE departments.dept_name = 'Sales' OR departments.dept_name = 'Development'
    ''', conn)
sales_dev_employees
```

```python
last_name_count=pd.read_sql(f'''
SELECT last_name, COUNT(last_name)
FROM employees
GROUP BY last_name
ORDER BY COUNT(last_name) DESC
    ''', conn)
last_name_count
```
